```python
from transformers import pipeline
import torch
import configparser
from diffusers import DiffusionPipeline
config_ini = configparser.ConfigParser()
config_ini.read('config.ini', encoding='utf-8')
```

```python
MOEL_NAME = config_ini["DEFAULT"]["model_name"]
CHECKPOINT = config_ini["DEFAULT"]["checkpoint"]
MODEL_PATH = f"./model/{MOEL_NAME}/{CHECKPOINT}/"
pipe = pipeline(
    "audio-classification",
    model=MODEL_PATH,
    torch_dtype=torch.float16,
    # torch_dtype=torch.bfloat16, 
    device=0,
)

def classify_audio(filepath):
    preds = pipe(filepath)
    outputs = {}
    for p in preds:
        outputs[p["label"]] = p["score"]
    return outputs
```

```python
FOLDER_NAME ="0_pingpong"
# FILE_NAME = "VSQSE_0060_pingpong_03-07.wav"
FILE_NAME = "ball_racket-02.wav"
# data_path = f"./data/test/{FOLDER_NAME}/{FILE_NAME}"
data_path = f"./data/{FILE_NAME}"
classify_audio(data_path)
```
