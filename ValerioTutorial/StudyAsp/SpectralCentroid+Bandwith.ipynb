```python
import librosa
import librosa.display 
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
debussy_file = "audio/debussy.wav"
redhot_file = "audio/redhot.wav"
duke_file = "audio/duke.wav"
debussy, sr = librosa.load(debussy_file)
redhot, _ =librosa.load(redhot_file)
duke, _ = librosa.load(duke_file)
```

```python
FRAME_SIZE = 1024
HOP_SIZE = 512

sc_debussy = librosa.feature.spectral_centroid(y=debussy, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
sc_redhot = librosa.feature.spectral_centroid(y=redhot, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
sc_duke = librosa.feature.spectral_centroid(y=duke, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
```

```python
sc_debussy.shape
```

## Visualize spectral centroid

```python
frames = range(len(sc_debussy))
t = librosa.frames_to_time(frames)
```

```python
plt.figure(figsize=(25,10))

plt.plot(t, sc_debussy, color='b')
plt.plot(t, sc_redhot, color='r')
plt.plot(t, sc_duke, color='y')

plt.show()
```

## Calculate bandwidth

```python
ban_debussy = librosa.feature.spectral_bandwidth(y=debussy, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
ban_redhot = librosa.feature.spectral_bandwidth(y=redhot, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
ban_duke = librosa.feature.spectral_bandwidth(y=duke, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
```

```python
ban_debussy.shape
```

## Visualize bandwidth

```python
plt.figure(figsize=(25,10))

plt.plot(t, ban_debussy, color='b')
plt.plot(t, ban_redhot, color='r')
plt.plot(t, ban_duke, color='y')

plt.show()
```
